```python
import pandas as pd
import json

df = pd.read_csv('/Users/dwanjiru/projects/kenya-coffee-research/data/processed/master_data_1990_2024.csv')

chart_data = df[['Temperature', 'Precipitation', 'Humidity', 'Solar_Radiation', 'Yield_Kg_Ha']].rename(columns={
    'Temperature': 'temperature',
    'Precipitation': 'precipitation',
    'Humidity': 'humidity',
    'Solar_Radiation': 'solar',
    'Yield_Kg_Ha': 'yield'
}).to_dict(orient='records')

with open('/Users/dwanjiru/projects/kenya-coffee-research/website/src/data/bivariateData.json', 'w') as f:
    json.dump(chart_data, f, indent=2)

print(f"Wrote {len(chart_data)} rows")
```
